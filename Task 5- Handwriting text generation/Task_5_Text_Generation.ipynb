```python
from datasets import load_dataset
dataset=load_dataset("Teklia/IAM-line")
print(dataset)
```

```python
texts=[item["text"] for item in dataset["train"]]
text=" ".join(texts)
text=text.lower()
print("Total characters:",len(text))
print(text[:500])
```

```python
chars=sorted(list(set(text)))
char_to_idx={c:i for i,c in enumerate(chars)}
idx_to_char={i:c for i,c in enumerate(chars)}
vocab_size=len(chars)
print("Vocabulary size:",vocab_size)
```

```python
import numpy as np
seq_length=40
X=[]
y=[]
for i in range(len(text)-seq_length):
    seq=text[i:i+seq_length]
    target=text[i+seq_length]
    X.append([char_to_idx[c] for c in seq])
    y.append(char_to_idx[target])
X=np.array(X)
y=np.array(y)
print("Training samples:",X.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
model=Sequential([Embedding(vocab_size,64,input_length=seq_length),LSTM(128,return_sequences=False),Dropout(0.2),Dense(vocab_size,activation="softmax")])
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(X,y,batch_size=128,epochs=30)
```

```python
def generate_text(seed,length=200):
    result=seed
    for _ in range(length):
        encoded=np.array([[char_to_idx[c] for c in seed]])
        pred=model.predict(encoded,verbose=0)[0]
        next_idx=np.random.choice(len(pred),p=pred)
        next_char=idx_to_char[next_idx]
        result+=next_char
        seed=seed[1:]+next_char
    return result
seed=text[100:140]
print(generate_text(seed))
```

```python
print(generate_text(text[200:240]))
print(generate_text(text[500:540]))
print(generate_text(text[1000:1040]))
```
